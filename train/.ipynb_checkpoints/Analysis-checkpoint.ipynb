```python
runningdf=[running1,running2,running3,running4,running5,running6,running7,running8,running9,running10]
runningdf
```

```python

#using low pass filter 
b, a = signal.butter(3, 0.1, btype='lowpass', analog=False)
walk_np=list(walkdf)
running_np=list(runningdf)
```

```python
low_passed_walk = []
for i in range(10):
    low_passed_walkX = list(signal.filtfilt(b, a, walk_np[i]["gFx"]))
    low_passed_walkY = list(signal.filtfilt(b, a, walk_np[i]["gFy"]))
    low_passed_walkZ = list(signal.filtfilt(b, a, walk_np[i]["gFz"]))
    low_passed_walk.append((pd.DataFrame(list(zip(low_passed_walkX, low_passed_walkY, low_passed_walkZ)),
                                            columns=["AccX", "AccY", "AccZ"])))
    
    
    
    
low_passed_running = []
for i in range(10):
    low_passed_runningX = list(signal.filtfilt(b, a, running_np[i]["gFx"]))
    low_passed_runningY = list(signal.filtfilt(b, a, running_np[i]["gFy"]))
    low_passed_runningZ = list(signal.filtfilt(b, a, running_np[i]["gFz"]))
    low_passed_running.append((pd.DataFrame(list(zip(low_passed_runningX, low_passed_runningY, low_passed_runningZ)),
                                            columns=["AccX", "AccY", "AccZ"])))
```

```python

########Walking data########
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='green') # cutoff frequency
plt.show()
```

```python
for i in range (10):
    fig, ax = plt.subplots(3,figsize=(15,21))
    
    fig.suptitle('Running reading no. {}'.format(i+1), fontsize=16)
    
    ax[0].plot(running_np[i]["gFx"], label='Original')
    ax[0].plot(low_passed_running[i]["AccX"], label='Filtered', linewidth=3.0)
    ax[0].legend(fontsize=20)
    ax[0].title.set_text('Acceleration X-axis')
    
    
    
    ax[1].plot(running_np[i]["gFx"], label='Original')
    ax[1].plot(low_passed_running[i]["AccX"], label='Filtered', linewidth=3.0)
    ax[1].legend(fontsize=20)
    ax[1].title.set_text('Acceleration Y-axis')
    
    
    
    
    ax[2].plot(running_np[i]["gFz"], label='Original')
    ax[2].plot(low_passed_running[i]["AccX"], label='Filtered', linewidth=3.0)
    ax[2].legend(fontsize=20)
    ax[2].title.set_text('Acceleration Z-axis')
    
    
    
 
```
